```
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import KFold, cross_val_score
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# StandardScaler DEMO
scaler = StandardScaler()
plt.plot(scaler.fit_transform(np.array([[100.0, 2.0], [50.0, 4.0], [70.0, 6.0]])));
```

```
train = pd.read_csv('train.csv', header=None)
test = pd.read_csv('test.csv', header=None)
Xtrn = train[[1,2]]
ytrn = train[0]
Xtst = test[[1,2]]
ytst = test[0]
```

```
cls = Perceptron(random_state=241)
cls.fit(Xtrn, ytrn)
predictions = cls.predict(Xtst)
accuracy = accuracy_score(ytst, predictions)
```

```
scaler = StandardScaler()
Xtrn_norm = scaler.fit_transform(Xtrn)
Xtst_norm = scaler.fit_transform(Xtst)
```

```
cls = Perceptron(random_state=241)
cls.fit(Xtrn_norm, ytrn)
predictions = cls.predict(Xtst_norm)
accuracy_norm = accuracy_score(ytst, predictions)
```
